# Python Libraries Used
***

### Flask (1.1.2)
*Source* : https://flask.palletsprojects.com/en/1.1.x/
<br>
*Use* : Provides the front end for Flaks and provides a GUI for web interactions. We are using a template from appseed (More documentation and credit provided in *Flask* documentation).

### Numpy (1.18.5)
*Source* : https://numpy.org/
<br>
*Use* : Powerful library that many other libraries are based on or intergrated with. Stands for *Numerical Python*. Used for manipulating arrays and dealing with NULL values.

### Pandas (1.1.4)
*Source* : https://pandas.pydata.org/
<br>
*Use* : Used extensively for DataFrames. Very powerful library for manipulating and storing data. 

### Pickle (1.1.2)
*Source* : https://docs.python.org/3/library/pickle.html
<br>
*Use* : Library that helps with saving and loading machine learning models. Used to save and load *Polynomial Regression* and *Random Forest*.


### SQLAlchemy (1.3.20)
*Source* : https://www.sqlalchemy.org/
<br>
*Use* : SQL Alchemy (or sqlite3) is the library used to interact with the SQL Lite database used by the code. It is responsible for turning strings/python code into SQL querries.

### Scikitlearn (0.23.2)
*Source* : https://sklearn.org/
<br>
*Use* : Open source machine learning library that provides tools for data mining and data analysis. Used for *Polynomial Regression* and *Random Forest* machine learning algorithums as well as various pre and post processing functions for the code.

### Talib (0.4.19)
*Source* : https://www.ta-lib.org/
<br>
*Use* : This library provides tools for technical stock analysis. We used this library to generate the following stock idicators : SAR, RSI, CCI, MACD, Bollanger Bands, EMA, Chaikin, WILLR. This library uses the data provided by yfinance and loaded into the sqlite database to preform its calculations.
<br>
*Note* : install with the following command | pip install TA-Lib

### Tensorflow (2.3.1)
*Source* : https://www.tensorflow.org/
<br>
*Use* : Open Source machine learning library made by google. Used primarily in this code with the Long short-term memory part of the project, a recurrent neural network machine learning algorithum. 

### Werkzeug (1.0.1)
*Source* : https://pypi.org/project/Werkzeug/
<br>
*Use* : WSGI (Web Server Gate Interface) library that works with Flask on the web intergration part of the project. (More documentation and credit provided in *Flask* documentation)

### yfinance (0.1.55)
*Source* : https://pypi.org/project/yfinance/
<br>
*Use* : Used to interact with Yahoo Finance API. Library extracts information based on a ticker value and presents the data to python in an DataFrame. We extract all basic information used in this program with this library.
